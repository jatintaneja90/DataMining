Question: A function that given a user id and movie id , returns a list of user ids for
other users that rated the movie identi_ed by the provided movie id with the same score.

```python
# this funciton will filter out instances where col value is valand return a new dataframe
def filterWantedAttribute(df,col,val):
    return df.loc[df[col] == val]
```

```python
# this funciton will filter out instances where col value is not val and return a new dataframe
def filterUnWantedAttribute(df,col,val):
    return df.loc[df[col] != val]
```

```python
def similar_rating_users(user_id, movie_id):
    if data.loc[data['movie_id'] == movie_id].empty:
        raise Exception("Given movie id %d is not available in data set"%movie_id)
    users_rated_given_movie = filterWantedAttribute(data,'movie_id', movie_id)
    if users_rated_given_movie.empty or users_rated_given_movie.loc[users_rated_given_movie['user_id'] == user_id].empty:
        raise Exception("Given user id %d is not available in data set"%user_id)
    desired_rating = filterWantedAttribute(users_rated_given_movie,'user_id', user_id)['rating'].item()
    given_user_filtered_df = filterUnWantedAttribute(users_rated_given_movie,'user_id', user_id)
    same_rating_users = filterWantedAttribute(given_user_filtered_df,'rating', desired_rating)
    return same_rating_users['user_id']
```
